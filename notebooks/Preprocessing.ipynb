```python
import pandas as pd
import numpy as np

from scipy.signal import savgol_filter
```

```python
data = pd.read_csv("/home/sonn/Sonn/Workspace/Projects/GANSpectralDataAugmentation/data/UV_Vis_NuocCam_ChuaDaoHam.csv")
```

```python
data.head(3)
```

Savgol Filter

```python
X = data.iloc[:, 4:]
```

```python
X = savgol_filter(X, window_length=5, polyorder=2, deriv=1)
```

```python
X_der = pd.DataFrame(X)
```

```python
y = data[['TenMau', 'DiaPhuong', 'NamLayMau', 'Giong']]
```

```python
X_der[['TenMau', 'DiaPhuong', 'NamLayMau', 'Giong']] = y
```
